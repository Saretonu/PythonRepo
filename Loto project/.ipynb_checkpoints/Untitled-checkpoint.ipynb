```python
f = open(pathNP, 'r')
l = f.readlines()
f.close()
```

```python
def listToString(s): 
    
    # initialize an empty string
    str1 = "" 
    
    # traverse in the string  
    for ele in s: 
        str1 += ele  
    
    # return string  
    return str1 
```

```python
r = listToString(l)
```

```python
def findDivStartAndEnd(className, strArray):
    response = [-1,-1]
    startFound = False
    divCount = 0
    startDivIndex = -1
    endDivIndex = -1
    RowIndex = -1
    for x in strArray:
        RowIndex +=1
        if startFound:
            if x.find('<div') >= 0:
                divCount += 1
            if x.find('</div>') >= 0:
                if divCount == 0:
                    endDivIndex = RowIndex
                else:
                    divCount -= 1
        else:
            if x.find('<div class="'+className+'">') >= 0:
                startDivIndex = RowIndex
                startFound = True
        if startDivIndex != -1 and endDivIndex != -1:
            response[0] = startDivIndex
            response[1] = endDivIndex
            findDivStartAndEnd = response
    response[0] = startDivIndex
    response[1] = endDivIndex
    findDivStartAndEnd = response
```

```python
res = [-1,-1]
print(len(res))
res = findDivStartAndEnd('game_results',l)

#print(res[0])
```
