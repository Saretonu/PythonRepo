```python
import requests

pageNumber = 1
Links_list = []
GameNr_list = []
htmlLink = "https://www.eestiloto.ee/osi/draws.do?gameSortType=1&sortProperty=DRAWDATE&sortDirection=false&pageNumber=" + str(pageNumber)
```

```python
while True :
    htmlLink = "https://www.eestiloto.ee/osi/draws.do?gameSortType=1&sortProperty=DRAWDATE&sortDirection=false&pageNumber=" + str(pageNumber)
    print(htmlLink)
    r = requests.get(htmlLink)
    
    R = str.strip(r.text)
    Index1 = R.find("main_table_layout sm-table-spacing")
    #print (Index1)

    start = 0
    stop = Index1
    # Remove charactes from index 5 to 10
    if len(R) > stop :
        R = R[0: start:] + R[stop + 1::]

    if Index1 != -1 :
        while True :

            Index2 = R.find("<tr>")
            #print (Index2)

            start = 0
            stop = Index2 + 21
            # Remove charactes from index 5 to 10
            if len(R) > stop :
                R = R[0: start:] + R[stop + 1::]

            Index3 = R.find('<a href="')
            #print (Index3)

            start = 0
            stop = Index3+8
            # Remove charactes from index 5 to 10
            if len(R) > stop :
                R = R[0: start:] + R[stop + 1::]

            Index4 = R.find('"')
            #print (Index4)

            start = 0
            stop = Index4
            # Remove charactes from index 5 to 10
            if len(R) > stop :
                line = R[0: Index4:]
                R = R[0: start:] + R[stop + 1::]

            #print (line)
            #print (R)

            if Index2 != -1 and Index3 != -1 and Index4 != -1 :
                Index5 = line.find("drawNumber=")
                if Index5 != -1 :
                    gameNr = line[Index5+11: line.find("&",Index5):] 
                    GameNr_list.append(gameNr)
                    Links_list.append(line)
            else :
                pageNumber += 1
                break
        else :
            print ("NOT FOUND")
    else :
        break
print (GameNr_list)
print (Links_list)
```
